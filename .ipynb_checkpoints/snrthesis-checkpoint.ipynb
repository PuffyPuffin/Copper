```python
import matplotlib.pyplot as plt
import numpy as np
from netCDF4 import Dataset as nc
import math
import statistics as stat

import cartopy.crs as ccrs
import cartopy.feature as cfeature
from cartopy.mpl.gridliner import LONGITUDE_FORMATTER, LATITUDE_FORMATTER
from collections import OrderedDict
from shapely.ops import cascaded_union
import itertools
from itertools import chain
```

```python
def lemon (file1, file2, graphname, plt1, plt2):

    ns = nc(file1)
    s = nc(file2)

    ns_reldepth = ns["cs"][:].data
    ns_lat = ns["lat"][:].data
    ns_lon = ns["lon"][:].data
    ns_depth = ns["h"][:].data
    
    ns_depthB = ns_reldepth < -0.99
    ns_ind = np.argmax(ns_depthB, axis = 0)

    ns_ind[ns_ind==0] = 720
    a = np.arange(0,1000)
    ns_latF = ns_lat[[ns_ind],[a]]
    ns_lonF = ns_lon[[ns_ind],[a]]
    ns_reldepthF = ns_reldepth[[ns_ind],[a]]
    ns_depthF = ns_depth[[ns_ind],[a]]

    s_reldepth = s["cs"][:].data
    s_lat = s["lat"][:].data
    print(s_lat[0,0])
    s_lon = s["lon"][:].data
    print(s_lon[0,0])
    s_depth = s["h"][:].data
    print(s_depth[0,0])
    s_depthB = s_reldepth < -0.99
    s_ind = np.argmax(s_depthB, axis = 0)
    s_ind[s_ind==0] = 720

    s_latF = s_lat[[s_ind],[a]]
    s_lonF = s_lon[[s_ind],[a]]
    s_depthF = s_depth[[s_ind],[a]]
    s_reldepthF = s_reldepth[[s_ind],[a]]
    

    labels = ["Copper", "Lead"]
    colors = ["darkorange","dimgrey"]
    ec = ["white", "k"]
    cu = abs(ns_reldepthF[0,:])
    pb = abs(s_reldepthF[0,:])
    data = [abs(ns_reldepthF[0,:]), abs(s_reldepthF[0,:])]
    bin1 = np.arange(1, 0, .02)
    print(bin1)
    
    
    fig = plt.figure()
    ax = fig.add_subplot(111)    # The big subplot  
    ax.set_xlabel("Number of Particles")
    ax.set_ylabel("Relative Depth (% of water column)")

    c,d,e = ax.hist([cu,pb], color = colors, label = labels, fill = True, orientation='horizontal', bins = 5)
    plt.setp(e[0], edgecolor='w', lw=1)
    plt.setp(e[1], edgecolor='k', lw=1)
    ax.set_xscale("log")
    ax.set_ylim(1,0)
    ax.legend(loc='right')

    plt.savefig(plt1)
    
    print(len(ns_ind[ns_ind==720])/len(ns_ind))
    print(len(s_ind[s_ind==720])/len(s_ind))
    
    def plotMap(locationDict, cmap='rainbow', figsize=[10,10]):

        plt.figure(num=5, figsize=figsize)  # create figure
        ax = plt.axes(projection=ccrs.PlateCarree())  # set map projection
    
        # add coastlines and axes ranges
        ax.coastlines(resolution='10m')
        ax.set_extent([ns_lonF.min()-.1, -122.2, 47.05, ns_latF.max()+.1])
        print(ns_latF.max(), ns_lonF.min())
        # add colored land and water (ocean feature has long loading time)
        

        
        ax.add_feature(cfeature.LAND, zorder=1, facecolor='burlywood')
        ax.add_feature(cfeature.OCEAN, zorder=1, facecolor= 'lightseagreen')
        rivers_NA = cfeature.NaturalEarthFeature('physical', 'rivers_north_america', '10m', edgecolor = 'lightseagreen', facecolor='none')
        ax.add_feature(rivers_NA, zorder=1)
        rivers = cfeature.NaturalEarthFeature('physical', 'rivers_lake_centerlines_scale_rank', '10m', edgecolor = 'paleturquoise', facecolor='none')
        ax.add_feature(rivers, zorder=1)
        islands = cfeature.NaturalEarthFeature('physical', 'minor_islands', '10m', facecolor='burlywood')
        ax.add_feature(islands, zorder=1)
        lakes = cfeature.NaturalEarthFeature('physical', 'lakes_north_america', '10m', facecolor='lightseagreen')
        ax.add_feature(lakes, zorder=1)
        playas = cfeature.NaturalEarthFeature('physical', 'playas', '10m', facecolor='paleturquoise', edgecolor = 'black')
        ax.add_feature(playas, zorder=1)
        plulakes = cfeature.NaturalEarthFeature('physical', 'lakes_pluvial', '10m', facecolor='paleturquoise', edgecolor = 'black')
        ax.add_feature(plulakes, zorder=1)
        
 
    
        # add title and gridlines, customize axes labels
        plt.title(graphname, fontsize='xx-large')
        gl = ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True,
                  linewidth=1, color='white', alpha=0.5, linestyle=':', zorder=2)
        gl.top_labels = False
        gl.xformatter = LONGITUDE_FORMATTER
        gl.yformatter = LATITUDE_FORMATTER
    
        # plot locations
        colors = ["darkorange", "dimgrey"]
        edgecolor = ["white", "black"]
        legendList = list(locationDict.keys())
        for i in range(len(locationDict)):  # for every location
            c = colors[i]  # get new color
            ec = edgecolor[i]
            coords = locationDict[legendList[i]]  # get coords from ordered dictionary
            lat = coords[0]
            lon = coords[1]
            plt.scatter(lon, lat, marker='o', color=c, zorder=3, label=legendList[i], edgecolor = ec)  # plot location
        
        ax.legend()  # add legend
        plt.savefig(plt2)
        plt.show()

    def main():

        # create and fill ordered location dictionary
        locationDict = OrderedDict()
        locationDict["Copper"] = [ns_latF, ns_lonF]
        locationDict["Lead"] = [s_latF, s_lonF]
        plotMap(locationDict)  # get map of Puget Sound showing data collection sites

    if __name__ == "__main__":
        main()
    def points2distance(start,end):
        start_latt = start[0]*math.pi/180
        end_latt = end[0]*math.pi/180
        d_latt = (end_latt - start_latt)
        start_long = start[1]*math.pi/180
        end_long = end[1]*math.pi/180
        d_long = (end_long - start_long)
        a = math.sin(d_latt/2)**2 + math.cos(start_latt) * math.cos(end_latt) * math.sin(d_long/2)**2
        c = 2 * math.asin(math.sqrt(a))
        return 6371 * c
    if __name__ == '__main__':
        d = np.zeros((2, 1000))
        for i in a:
            ns_start = (ns_lat[0,i],ns_lon[0,i])
            ns_end = (ns_latF[0,i],ns_lonF[0,i])
            d[0,i] = points2distance(ns_start,ns_end)
    if __name__ == '__main__':
        for i in a:
            s_start = (s_lat[0,i],s_lon[0,i])
            s_end = (s_latF[0,i],s_lonF[0,i])
            d[1,i] = points2distance(s_start,s_end)

        return d, data
```
